```python
first_innings = match_df.merge(first_dataset,on='match_number')
first_innings
```

```python
first_innings = first_innings[first_innings['innings']==1]
first_innings.shape
```

```python
first_innings = first_innings.drop_duplicates(subset=['match_number', 'city', 'winner', 'team1', 'team2', 'match_id', 'innings', 'ball', 'batting_team', 'bowling_team', 'runs_off_bat', 'extras', 'total_run_y'])
```

```python
# Group by 'match_number' and find the maximum 'total_run_x' for each match number
max_total_run_x_per_match = first_innings.groupby('match_number')['total_run_x'].max()

# Convert the result to a list
result_list = max_total_run_x_per_match.reset_index().values.tolist()
```

```python
result_list
```

```python
# Convert the result list to a DataFrame with columns 'match_number' and 'max_total_run_x'
target_df = pd.DataFrame(result_list, columns=['match_number', 'max_total_run_x'])

# Merge 'second_inning' with the target DataFrame based on 'match_number'
second_inning = pd.merge(second_inning, target_df, on='match_number', how='left')
```

```python
second_inning
```

```python
# Convert the 'ball' column to string and split it into two columns
second_inning[['over', 'balls']] = second_inning['ball'].astype(str).str.split('.', expand=True)

# Convert 'over' and 'balls' columns to integers
second_inning['over'] = second_inning['over'].astype(int)
second_inning['balls'] = second_inning['balls'].astype(int)
```

```python
second_inning = second_inning.drop(['ball'], axis=1)
second_inning
```

```python
second_inning['current_Score'] = second_inning.groupby('match_number').cumsum()['total_run_y']
second_inning
```

```python
second_inning['runs_left'] = second_inning['max_total_run_x'] - second_inning['current_Score']
second_inning
```

```python
second_inning['balls_left'] = 126 - (second_inning['over']*6 +second_inning['balls'])
second_inning
```

```python
second_inning.columns
```

```python
# Select the columns of interest
columns_to_check = ['current_Score', 'runs_left', 'balls_left']

# Check how many values in each column are negative
negative_counts = second_inning[columns_to_check].lt(0).sum()

# Display the counts
print("Number of negative values in each column:")
print(negative_counts)
```

```python
second_inning['player_dismissed'] = second_inning['player_dismissed'].fillna("0")
second_inning['player_dismissed'] = second_inning['player_dismissed'].apply(lambda x:x if x=="0" else "1")
second_inning['player_dismissed'] = second_inning['player_dismissed'].astype('int')
wickets = second_inning.groupby('match_number').cumsum()['player_dismissed']
second_inning['wickets_remaining'] = 10-wickets
second_inning
```

```python
second_inning['crr'] = round((second_inning['current_Score']*6)/(120 - second_inning['balls_left']),2)
second_inning['rrr'] = round((second_inning['runs_left']*6)/(second_inning['balls_left']),2)
second_inning
```

```python
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
second_inning['results'] = second_inning.apply(winner,axis=1)
second_inning.head()
```

```python
required_data = second_inning[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','max_total_run_x','crr','rrr','results']]
required_data
```

```python
required_data = required_data.replace([np.inf, -np.inf], np.nan)
required_data = required_data.dropna()
```

```python
required_data = required_data[required_data['crr'] >= 0]
```

```python
columns_to_check = ['runs_left', 'balls_left', 'wickets_remaining', 'max_total_run_x', 'crr', 'rrr', 'results']

required_data = required_data[~(required_data[columns_to_check] < 0).any(axis=1)]
```

```python
numeric = required_data[['runs_left', 'balls_left', 'wickets_remaining', 'max_total_run_x', 'crr', 'rrr', 'results']]
has_infinite_values = np.any(np.isinf(numeric))

# Check for negative values in the entire DataFrame
has_negative_values = (numeric < 0).any().any()

# Check for null values in the entire DataFrame
has_null_values = numeric.isnull().any().any()

# Display the results
print(f"Has Infinite Values: {has_infinite_values}")
print(f"Has Negative Values: {has_negative_values}")
print(f"Has Null Values: {has_null_values}")
```
